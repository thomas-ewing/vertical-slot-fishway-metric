```python
import matplotlib.pyplot as plt
import numpy as np
import verticalslot_metric as vm
```

```python
display_fish_size_array = np.arange(20,1501,5)
```

```python
# TDR

display_obs_TDR = np.arange(0,175,1)
turb_tf_data = np.zeros([len(display_obs_TDR), len(display_fish_size_array)])

ari_criteria_size = np.array([[20,100],[100,200],[200,700],[700,1200]])
ari_criteria_value = np.array([25,30,50,50])

for i in range(len(display_obs_TDR)):
    for j in range(len(display_fish_size_array)):
        turb_tf_data[i,j] = vm.calc_turb_metric(display_fish_size_array[j], display_obs_TDR[i])

fig = plt.figure(0, figsize=(12, 5))
fig.set_facecolor('lightgrey')
plt.pcolor(display_fish_size_array, display_obs_TDR, turb_tf_data, cmap='PiYG', vmin=0, vmax=1)

plt.xlabel('Fish Size (mm)')
plt.ylabel('Turbulent Dissipation Rate ($Wm^{-3}$)')

cbar = plt.colorbar()
cbar.set_label('Transfer Function Output')

# Plot ARI Design Criteria

ha_list = ["left", "left", "left", "left"]
va_list = ["center", "center", "center", "center"]
offset_x = [25,50,25,25]
offset_y = [200,-25,25,-25]

for ari_level in range(len(ari_criteria_size)):

    plt.plot(ari_criteria_size[ari_level], [ari_criteria_value[ari_level],ari_criteria_value[ari_level]],'ko--')

    label = '{} to {} mm\nMax. TDR: {} $Wm^{{-3}}$\n(O\'Connor et al., 2017)'.format(ari_criteria_size[ari_level][0],
        ari_criteria_size[ari_level][1], ari_criteria_value[ari_level])

    bbox = dict(facecolor='white', alpha=0.5, edgecolor='None', boxstyle='square,pad=0.5')
    
    plt.annotate(label,
                 xy=[(ari_criteria_size[ari_level][0]+ari_criteria_size[ari_level][1])/2, ari_criteria_value[ari_level]],
                 xytext=(offset_x[ari_level],offset_y[ari_level]), textcoords="offset points",
                 horizontalalignment=ha_list[ari_level], verticalalignment=va_list[ari_level],
                 bbox=bbox, size=8, arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=90,rad=10"))

plt.savefig('TDR_TF.png', dpi=300, bbox_inches='tight',)
plt.show()
```
